**Operators**

```python
a=15
b=4

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)
```

**A simple calci capable of performing addition, substraction, multiplication and division.**

```python
c=4
d=5
print("The values for the given calculater are:\n",c, "and", d)
print("The addition for c and d is:\t",c+d)
print("The substraction for c and d is:\t", c-d)
print("The multiplication for c and d is:\t", c*d)
print("The division for c and d is:\t", c/d)
```
