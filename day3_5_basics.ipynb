**A simple calci capable of performing addition, substraction, multiplication and division.**

```python
c=4
d=5
print("The values for the given calculater are:\n","\"",c,"\"", "and", "\"",d,"\"")

print("The Addition for c and d is:",c+d)
print("The Substraction for c and d is:", c-d)
print("The Multiplication for c and d is:", c*d)
print("The Division for c and d is:", c/d)
print("The Modulus for c and d is:", c%d)
print("The Exponential for c and d is:", c**d)
print("The Floor division for c and d is:", c//d)
print("The addition is: ", "\"", c+d, "\"", sep="")
```

**Operators**

```python
a=15
b=4

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)
```
