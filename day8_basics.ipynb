**upper():**

```python
str1="KirTImaN"
print(str1.upper())
```

**lower():**

```python
print(str1.lower())
```

**strip():**

```python
str2=" Silver Spoon"
print(str2.strip())
```

**rstrip():**

```python
str3="!!!Hello!!!"
print(str3.rstrip("!"))
```

**replace():**

```python
print(str2.replace("Sp", "M"))
```

**split():**

```python
print(str2.split(" "))
```

**Capitalize():**

```python
str4="hello"
# capstr1=str4.capitalize()
print(str4.capitalize())
# print(capstr1)
str5="hello World"
print(str5.capitalize())
```

**center():**

```python
str45="Welcome to the Console!!!"
print(str45.center(50, "_"))
```

**count():**

```python
sss="Where are you going bro?"
countsss=sss.count("o")
print(countsss)
```

**endswith():**

```python
css=sss.endswith("?")
print(css)
print(sss.endswith("are",3, 9))
print(sss.endswith("are",3, 13))
```

**find():**

```python
tits="Shall we end it here for today?"
print(tits.find("it"))
print(tits.find("pops"))
```

**index():**

```python
print(tits.index("here"))
print(tits.index("Kirtiman")) # will throw an error
```
