```
import time
current_hour=int(time.strftime("%H"))

user_role=input("what's your role:\t")
match user_role:
  case "admin":
    print("Full access.")
  case "user":
    if (current_hour < 18):
      print("Limited Access")
    else:
      print("Access expired for today")
  case _:
    print("Check your user name and try again\nThank You...")
```

```
import re

needed_storage_input = input("How much storage do you need (e.g., 10GB, 500MB, 1TB):\t")

# Define conversion factors to GB
conversion_factors = {
    'MB': 0.001,  # 1 MB = 0.001 GB
    'GB': 1.0,    # 1 GB = 1 GB
    'TB': 1000.0  # 1 TB = 1000 GB
}

# Use regex to extract the number and the unit
# This regex looks for: a number (\d+), optional space (\s*), then letters (\w+)
match = re.match(r'(\d+)\s*([a-zA-Z]+)', needed_storage_input)

if match:
    numeric_value = int(match.group(1)) # The number part
    unit = match.group(2).upper()      # The unit part, converted to uppercase for consistency

    if unit in conversion_factors:
        # Convert the input to a standardized unit (e.g., GB)
        needed_storage_gb = numeric_value * conversion_factors[unit]
        print(f"You entered: {numeric_value} {unit}")
        print(f"This is equivalent to: {needed_storage_gb:.2f} GB")

        # Example comparison (you can replace 'available_storage_gb' with your actual value)
        available_storage_gb = 50.0 # Let's assume you have 50 GB available
        if needed_storage_gb <= available_storage_gb:
            print(f"You need {needed_storage_gb:.2f} GB. Available storage: {available_storage_gb:.2f} GB. Storage is sufficient.")
        else:
            print(f"You need {needed_storage_gb:.2f} GB. Available storage: {available_storage_gb:.2f} GB. Insufficient storage.")

    else:
        print(f"Error: Unknown unit '{unit}'. Please use MB, GB, or TB.")
else:
    print("Error: Invalid input format. Please enter a number followed by a unit (e.g., 10GB, 500MB).")
```

Let's look at an example of how to handle mixed input:

```
user_input = input("Please enter a number or a string with numbers:\t")

try:
    # Try to convert the input to an integer
    numeric_value = int(user_input)
    print(f"You entered a valid integer: {numeric_value}")
except ValueError:
    # If it fails, it means the input was not purely an integer
    print(f"Your input '{user_input}' could not be converted to a whole number directly.")
    # You can then process it as a string
    print(f"Processing as a string: {user_input.upper()}")

    # If you need to extract numbers from a mixed string (e.g., '10GB')
    import re
    numbers_found = re.findall(r'\d+', user_input)
    if numbers_found:
        extracted_number = int(numbers_found[0]) # Takes the first number found
        print(f"Found a number in your input: {extracted_number}")
    else:
        print("No numbers found in your input string.")
```
