**Small Arthemetic operation using input()**

```python
c=int(input("Please enter the first number:\t"))
d=int(input("Please enter the second number:\t"))
print("The addition of both numbers is:", c+d)
print("The subtraction of both numbers is:", c-d)
print("The division of both numbers is:", c/d)
print("The multiplication of both numbers is:", c*d)
print("The floor division of both numbers is:", c//d)
```

**Corrected Version**

```python
q=input("how many GB storage would u like to have: ")
GB=float(q)
storage_cost=0.2*GB
tax_amount=storage_cost*0.10
final_bill=storage_cost+tax_amount
print(f"""
---Azzure Bill Storage---
Storage Cost: ${storage_cost}
Tax(10%): ${tax_amount}
Total: ${final_bill}
""")
```
