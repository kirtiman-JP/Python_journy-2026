```python
# Prompt the user to enter the desired storage in GB and convert it to a floating-point number.
needed_GB = float(input("How many GB would you like to have:"))

# Prompt the user to enter the discount percentage and convert it to a floating-point number.
# The input discount should be divided by 100 to be treated as a percentage.
asked_disc = float(input("Discount: ")) / 100

# Calculate the base cost of storage by multiplying the needed GB by the cost per GB (0.5).
base_cost = needed_GB * 0.5

# Calculate the discount amount by multiplying the base cost by the asked discount percentage.
discount_amt = base_cost * asked_disc

# Calculate the total cost after applying the discount.
total = base_cost - discount_amt

# Print a formatted bill summary showing the base cost, discount, and total cost.
print(f"""
---AZURE BILL---
Base Cost: ${base_cost})
Discount: ${discount_amt}
Total: ${total}
""")
```
