**Match case statements**

```python
com = "quit"
match com:
  case "start":
    print("System starting....")
  case "stop":
    print("System stopping....")
  case _:
    print("Unknown command....")
```

```python
we=int(input("plese enter the type of error u want to find about?\t"))
match we:
  case 200:
    print("OK")
  case 404:
    print("Not found")
  case 500:
    print("Server error")
  case _:
    print("Maybe you should check your input again....")
  #Here i had encontered an error that i compared command wich was other varable
  #so whatever coms after match is the variable we want to check here the variable was we so wee compare (match we:)
```

```python
see=int(input("plese enter the type of error u want to find about?\t"))
match see:
  case 200 | 201 | 204:
    print("Request successful!")
  case _:
    print("Maybe you should check your input again....")
# , means this this this like and but pipe (|) means or so 200, 201, 204 could mean this and this and this but 200 | 201 | 204 means this or this or this
```

```python
command=input("Please enter the command\t")
match command:
  case "start":
    print("System is powering on....")
  case "stop":
    print("System is shutting down....")
  case "restart":
    print("System is rebooting")
  case _:
    print("Error: command not recognised....")
```
