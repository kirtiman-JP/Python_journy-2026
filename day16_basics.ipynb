**Required Arguments**

```
#THESE ARE THE "MUST_HAVES" THE FUNCTION CANNOT RUN WITHOUT THEM
#eg. to turn on a smart light but need to specify wich room
def turn_on_light(room):#argument is a box for input data. O r simple words its an temporary variable.
  if room == "kitchen":
   print(f"Lighting up the\t{room}!")
  elif room == "washroom":
    print(f"Lighting up the\t{room}")
  else:
    print("Try input again")


turn_on_light("washroom")
# turn_on_light("kitchen")
```

**Default Arguments**

```
#U PROVIDE AN FEEDBACK OR DEFAULT VALUE IN CASE THE USER FORGETS TO GIVE ONE
#eg. if u dont define the brightness it returns to it's defsult value
def brightnes_ss(room, brightness=100):
  print(f"{room} light set to {brightness}%")

brightnes_ss("kitchen!")#uses default 100%
brightnes_ss("kitchen", 50)#overrides default to 50
```

**Keyword Arguments (named)**

```
#YOU CALL THE FUNCTION BY NAMING THE PARAMETERS. THIS MEANS THE ORDER DOSN'T  MATTER.
#eg. Telling the House: "Set colour to Blue and room to bath"
def colour_light(room, colour):
  print(f"setting {room} to {colour}")

colour_light(colour="blue", room="washroom") #ordr swapped but it works

def try_try(vehicle, brand):
  print(f"i like {vehicle} of {brand}'s ")

try_try(brand="Lamborgini", vehicle="Car")
```

**VARIABLE ARGUMENTS (\*args)**

```
#THIS ALLOWS A FUNCTION TO TAKE ANY NUMBER OF POSITIONAL ARGUMENTS. IT STORES THEM IN TUPLE.
#eg.A "shopping list" where you don't know how many items you will add
def add_to_cart(*items):
  for item in items:
    print(f"Added {item} to cart.")

add_to_cart("Milk", "Eggs", "Sugar", "Bread", "Jams(Blueberry, mango)", 556)
```

**KEYWORD ARBITRARY ARGUMENTS (\*\*kwargs)**

```
#THIS TAKES ANY NUMBER OF NAMED ARGUMENTS AND STORES THEM IN A DICTIONARY.
#eg.a "user profile" where you can add extra details like age, city, or job.
def save_user(**details):
  print(details) #{'name':'Alex', 'role':'Dev'}+

save_user(names="Alex", role="Dev")
```
