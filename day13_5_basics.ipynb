**Interrupted finish / Break**

```python
count=1
while count <=3:
  if count==2:
    break #Quitting early
  print(f"Lap{count}")
  count=count+1
else:
  print("This willl not print")
#u can do if statments in while loops
#so as discussed if break is used and the loop dosn't end naturally
#the else statement won't print
```

**Natural finish**

```python
energy=1
while energy<3:
  print(f"Running lap{energy}")
  energy = energy+1
else:
  print("Final lap done.\nRace completed")
```

**Emulating do while loop**

```python
#The "one free sample" example
while True:
  choice=input("Do you want a cookie? (yes/no)")
  #we do work first then check the condition to break
  if choice =="no":
    break
  print("Here is the cookie")
```
