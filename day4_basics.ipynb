**TypeCasting**


**Explict Type casting**


```python
string = "15"
number = 7
string_number = int(string)
sum = number + string_number
print("The sum of both number is: ", sum)
```

**Implict Type casting**

```python
a=7
print("The type value of a is:",type(a))
b=3.0
print("The type value of b is:",type(b))
c=a+b
print("The type value of c is:", type(c))
print(c)
```

```python
print(float("1.5"))
print(int(True), int(False))
```

```python
age = 25
age=str(age)
print("I am\t"+ str(age))
```

```python
#Simulated user input
user_input_1 = "10"
user_input_2="20"
#Explict type casting
val1=int(user_input_1)
val2=int(user_input_2)
print(f"Total: {val1 + val2}")
```

**F string (Formatted string)**

```python
server="az-01"
cores=8
print(f"The server is\t{server} and the core is\t{cores}")
```

```python
price=100
tax=0.18
print(f"Total Price with Tax: {price+(price*tax)}")
```

```python
stored_value="500.50"
#try to convert it directly to int()  it will crash
#practice the "double cast":
final_value=(float(stored_value))
print(f"the rounded value is: {final_value}")
print(f"the decimal value is: {int(final_value)}")
```
