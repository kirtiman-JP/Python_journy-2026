```python
import time
current_hour =int(time.strftime('%H')) # Get the current hour as an integer.
# error came because you tried to compare the time module itself (which is like a toolbox) directly with a number.
# Python needs a specific number (like the current hour) to make that comparison.
#u only compare here with %H cause %M and%s would throw error.
 #cause strftime chnges the string to int type so if its 14:04:60 then error cause of:
if(current_hour>= 5 and current_hour < 12):
  print("Good morning Sir")
elif(current_hour >= 12 and current_hour < 17):
  print("Good evening sir")
else:
  print("Good night sir")
```

```python
#.strftime is an function that gives us hour and time from the lib
import time
timestamp=time.strftime('%H:%M:%S') #The reason we use : in b/w is if we direct print %H%M%Sit will stick to to feel it like real clock its done
print(timestamp)
timestamp=time.strftime('%H')
print(timestamp)
timestamp=time.strftime("%M")
print(timestamp)
timestamp=time.strftime("%S")
print(timestamp)
```
