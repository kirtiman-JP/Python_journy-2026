**while with else**

```
cherry=10                    #SUBSTRACTION: you v should be bigger then given num v>10
print("READY TO LAUNCH IN:") #ADDITION: v should be smaller then given num v<10
while (cherry > 0):
  print(f"T-{cherry}")
  cherry=cherry-1
else:
  print("T-0\nLaunching successful")
```

```
#Natural finish
#An else in while only runs when the code ends naturally ends with the false or the loop ending
#butif there is a break in the code then the else won't work
count=1
while count <=3:
  print(f"lap{count}")
  count =  count+1
else:
  print("Race finished naturally!")
```
