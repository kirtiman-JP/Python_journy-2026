Emulating DO-while loop syntax

```
while True:
  #1. The code runs here first (The "Do" part)
  #2. Check the condition at the bottom
  if exit_condition:
    break
```

```
while True:
  password = input("Enter Your password:\t") #This runs first (The "DO")

  if password == "heloo123":
   print("Access granted")
   break
  else:
    print("Access denied try again")
#THink of it like vending machine. You put the coin in first (DO)
#Then the machine checksif uu have enough money. If you do it gives your desired drink and stops (break)
#If not it either waits for more coins or returns it back
```

Break statement (Stop)

```
#We use break statement when we need to stop the entire loop cause either we found somthing we were looking for or got
#an critical error
#EG For scanning 10 servers
for server_id in range(1, 11):
  print(f"Checking Server {server_id}...")

  if server_id==5:
    print("Virus found".upper(), "Emergency stop initiated")
    break
#U can think of it also like. Time complexity is the worst case.
  #if the virus is at end u r done for
#But if we find it earlier or on avg place we can end it so that it elps us exit early
```

Continue statemnt (Skip)


```
#We just use this when we want to skip the rest of thecode in current loop u want
 #and jump straight to next one
#Scanning 10 servers for updates

for server_id in range(1,11):
  if server_id==3:
   print("Server 3 is offline skipping to next one..") #continue will atleat print the code above it if u wanted the
                                                       #code to be run but put it below then it will skip it.
   continue#Skip the print line blow or code and go to server4
  print(f"Scanning Server {server_id} for updates")
```

```
servers=["Server_A", "OFFLINE", "Server_C", "VIRUS_DEDUCTED", "server_E"]

for s in servers:
  if s=="OFFLINE":
    print("Skipping maintainance")
    continue #Skip "Server_B" and go to "server_C"
  if s == "VIRUS_DEDUCTED":
    print("Critical error.".upper(), "Virus found!, Shutting down the scan.")
    break #Stop the entire loop we arn't looking at "server_E"
  print(f"performing maintainance on {s}...")
```
