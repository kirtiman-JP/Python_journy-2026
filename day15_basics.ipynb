```
#Define the function and its input 'mailbox' (parameter)
def find_error(log_list):
  #Start the loop.
  #'enumerate' gives us two things at once:
  #1. The position (index)
  #2. The actual word (status)
  for index, status in enumerate(log_list):
    #Check if the current word is our target
    if status == "CRITICAL":
      #If found , print the location
      print(f"Found error at position: {index}")
      #'return' sends the index back and STOPS the function immediately.
      #This replaces the need for 'break'!
      return index
  #This only runs if the loop finishes without finding "CRITICAL"
  return -1

#1.Define the data
my_logs=["OK", "WAITING", "CRITICAL", "OK"]
#2.Call the function and PASS the data
found_at=find_error(my_logs)
```
