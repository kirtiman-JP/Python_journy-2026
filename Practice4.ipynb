nesting an if. if in if condition


```
total_score=int(input("Enter your total score:\t"))
interview_score=int(input("Enter your interview score\t"))
if total_score > 80:
  if interview_score >90:#Nestedright away
    print("You got the job")
  else:
    print("Great score, but keep practicing interviews.")
else:
  print('Maybe try again later. Better by next time...')
```
